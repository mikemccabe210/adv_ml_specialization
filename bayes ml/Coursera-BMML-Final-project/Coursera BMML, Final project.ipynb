## Model description
We will first train variational autoencoder on face images to compress them to low dimension. One important feature of VAE is that constructed latent space is in some sence continous. That means, that we can safely move along the space and any point along our path will be reconstructed into a valid face.

Using this continuous latent space we can use bayesian optimization to maximise some similarity function between person's face in victim/witness's memory and a face reconstructed from current point of latent space. Bayesian optimization is an appropriate choice here, since people start to forget detales about the attacker after they were shown many similar photos. Because of this, we want to reconstruct the photo with the smallest possible number of trials.

## Search procedure

Now that we have a way to reconstruct images, we need to set up an optimization procedure to find a person that will be the most similar to the one we are thinking about. To do so, we need to set up some scoring utility. Imagine that you want to generate an image of Brad Pitt. You start with a small number of random samples, say 5, and rank them according their similarity to your vision of Brad Pitt: 1 for the worst, 5 for the best. You then rate image by image using GPyOpt that works in a latent space of VAE. For the new image you need to somehow assign a real number that will show how good this image is. The simple idea is to ask a user to compare a new image with previous images (along with their scores). User then enters score to a current image.

````python
class FacialComposit:
    def __init__(self, decoder, latent_size):
        self.latent_size = latent_size
        self.latent_placeholder = tf.placeholder(tf.float32, (1, latent_size))
        self.decode = decoder(self.latent_placeholder)
        self.samples = None
        self.images = None
        self.rating = None

    def _get_image(self, latent):
        img = sess.run(self.decode, 
                       feed_dict={self.latent_placeholder: latent[None, :]})[0]
        img = np.clip(img, 0, 1)
        return img

    @staticmethod
    def _show_images(images, titles):
        assert len(images) == len(titles)
        clear_output()
        
        n = len(titles)
        plt.figure(figsize=(3*3, 3*int(np.ceil(n/4))))
        for i in range(n):
            plt.subplot(int(np.ceil(n/4)), 4, i+1)
            plt.imshow(images[i])
            plt.title(str(titles[i]))
            plt.axis('off')
        plt.show()

    @staticmethod
    def _draw_border(image, w=2):
        bordred_image = image.copy()
        bordred_image[:, :w] = [1, 0, 0]
        bordred_image[:, -w:] = [1, 0, 0]
        bordred_image[:w, :] = [1, 0, 0]
        bordred_image[-w:, :] = [1, 0, 0]
        return bordred_image

    def query_initial(self, n_start=5, select_top=None):
        '''
        Creates initial points for Bayesian optimization
        Generate *n_start* random images and asks user to rank them.
        Gives maximum score to the best image and minimum to the worst.
        :param n_start: number of images to rank initialy.
        :param select_top: number of images to keep
        '''
        if select_top:
            pass
        else:
            select_top = n_start
        self.samples = np.zeros((select_top, self.latent_size))### YOUR CODE HERE (size: select_top x 64 x 64 x 3)
        self.images = np.zeros((select_top, 64, 64, 3)) ### YOUR CODE HERE (size: select_top x 64 x 64 x 3)
        self.rating = np.zeros(select_top) ### YOUR CODE HERE (size: select_top)
        
        ### YOUR CODE:
        ### Show user some samples (hint: use self._get_image and input())
        samples = np.random.randn(n_start, self.latent_size)
        images = [self._get_image(sample) for sample in samples]
        labels = ['Image %s' % i for i in range(n_start)]
        self._show_images(images, labels)
        rating = np.zeros(n_start)
        for i in range(n_start):
            rating[i] = float(input('Rate %s (1-10) on how well it fits the requirements?\n' % labels[i]))
        top_args = np.argsort(-1 * self.rating)[:select_top]
        for i, index in enumerate(top_args):
            self.rating[i] = rating[index]
            self.images[i] = images[index]
            self.samples[i] = samples[index]

        # Check that tensor sizes are correct
        np.testing.assert_equal(self.rating.shape, [select_top])
        np.testing.assert_equal(self.images.shape, [select_top, 64, 64, 3])
        np.testing.assert_equal(self.samples.shape, [select_top, self.latent_size])

    def evaluate(self, candidate):
        '''
        Queries candidate vs known image set.
        Adds candidate into images pool.
        :param candidate: latent vector of size 1xlatent_size
        '''
        initial_size = len(self.images)
        
        ### YOUR CODE HERE
        ## Show user an image and ask to assign score to it.
        ## You may want to show some images to user along with their scores
        ## You should also save candidate, corresponding image and rating
        top_n = min(3, initial_size)
        current_top_3 = np.argsort(-1 * self.rating)[:top_n]
        print('Top Rated')
        
        new_candidate = np.random.randn(self.latent_size)
        cand_image = self._get_image(new_candidate)
        self._show_images(self.images[current_top_3].tolist() + [cand_image],
                          self.rating[current_top_3].tolist() + ['New Image'])
#         self._show_images([cand_image], ['New Image'])
        candidate_rating = float(input('Rate new image? (1-10)'))### YOUR CODE HERE
        
        self.images = np.vstack([self.images, cand_image[np.newaxis, :, :, :]])
        self.rating = np.hstack([self.rating, np.array([candidate_rating])])
        self.samples = np.vstack([self.samples, new_candidate[np.newaxis, :]])
        
        assert len(self.images) == initial_size + 1
        assert len(self.rating) == initial_size + 1
        assert len(self.samples) == initial_size + 1
        return candidate_rating

    def optimize(self, n_iter=10, w=4, acquisition_type='MPI', acquisition_par=0.3, n_start = 5):
        if self.samples is None:
            self.query_initial(n_start = n_start)

        bounds = [{'name': 'z_{0:03d}'.format(i),
                   'type': 'continuous',
                   'domain': (-w, w)} 
                  for i in range(self.latent_size)]
        optimizer = GPyOpt.methods.BayesianOptimization(f=self.evaluate, domain=bounds,
                                                        acquisition_type = acquisition_type,
                                                        acquisition_par = acquisition_par,
                                                        exact_eval=False, # Since we are not sure
                                                        model_type='GP',
                                                        X=self.samples,
                                                        Y=self.rating[:, None],
                                                        maximize=True)
        optimizer.run_optimization(max_iter=n_iter, eps=-1)

    def get_best(self):
        index_best = np.argmax(self.rating)
        return self.images[index_best]

    def draw_best(self, title=''):
        index_best = np.argmax(self.rating)
        image = self.images[index_best]
        plt.imshow(image)
        plt.title(title)
        plt.axis('off')
        plt.show()
````

The algorithm used is essentially the same as the vanilla algorithm described. 

### Initial Pass
First, users are presented with a set of n images (default = 5). These are generated by sampling from the latent space using a standard normal gaussian. The latent samples are feed through the decoder to generate images. The users are prompted to score each of these images in sequence according to some input criteria. This input criteria does not need to be explicitly specified.

These initial scores form a set of seed rows for our GP optimization where our X values are our latent space samples and y values are our ratings. These are used to estimate the rating and the variance of that rating at each step. 

### Iterative Improvement
The process runs for a parameterized number of steps. At each step, the GP-based optimization explores the latent space according to a configurable acquisition function (default is MPI) in order to maximize the rating of the image. MPI in particular looks for the point with the highest probability of exceeding the current best prediction, though other options use different approaches.

Once the new latent value for the step is selected by GPyOpt, the latent vector is passed through the decoder and the resulting image presented to the end user for scoring. The top 3 scoring images thus far in the optimization are also fed to the end user for reference. When the user rates the new image, the image, sample, and rating are fed back into the GP used for optimization to continue improving rating estimates. 

### Limitations
- Using the pretrained model from CelebA, it looks like there's a bias towards certain facial characteristics. Glasses and facial hair are rare and there is little variety in male facial structure. This makes it difficult to optimize for certain features.
- It is difficult to optimize over several features simultaneously due to the limited variety in scores. Optimizing for hair color and facial hair, two images may achieve the same rank because they each do well on one criteria and poorly optimizing over the other. 
- Along with the above issue, the reliance on human scoring makes the process tedious. It can take a large number of evaluations to converge.
